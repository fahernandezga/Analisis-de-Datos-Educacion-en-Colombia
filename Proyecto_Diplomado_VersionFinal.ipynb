Primero realizamos una copia del dataframe que será la que usaremos para el análisis y en la que realizaremos los cambios respectivos. Además eliminaremos las variables categóricas y numéricas que no son importantes para nuestro análisis, como lo observamos en el análisis exploratorio de los datos

```python
df = estadisticas_MEN.copy()
df = df.drop(['CÓDIGO_MUNICIPIO', 'CÓDIGO_DEPARTAMENTO', 'CÓDIGO_ETC', 'ETC', 'POBLACIÓN_5_16', 'TAMAÑO_PROMEDIO_DE_GRUPO', 'SEDES_CONECTADAS_A_INTERNET'], axis=1)
```

Realizamos la corrección de las variables categóricas

```python
df['MUNICIPIO'] = df['MUNICIPIO'].replace('Bogotá, D.C.', 'Bogotá')
df['MUNICIPIO'] = df['MUNICIPIO'].replace('Bogotá D.C.', 'Bogotá')
```

```python
municipios = df['MUNICIPIO']
municipios
```

```python
df['AÑO'] = df['AÑO'].replace('2,021', '2021')
```

```python
anios = df['AÑO'].unique()
anios
```

```python
df['DEPARTAMENTO'] = df['DEPARTAMENTO'].replace('Archipiélago de San Andrés, Providencia y Santa Catalina', 'San Andrés y Providencia')
df['DEPARTAMENTO'] = df['DEPARTAMENTO'].replace('Archipiélago de San Andrés. Providencia y Santa Catalina', 'San Andrés y Providencia')
```

```python
df['DEPARTAMENTO'] = df['DEPARTAMENTO'].replace('Bogotá, D.C.', 'Bogota, D.C.')
df['DEPARTAMENTO'] = df['DEPARTAMENTO'].replace('Bogotá D.C.', 'Bogota, D.C.')
```

```python
departamentos = df['DEPARTAMENTO'].unique()
departamentos
```

### **3.1.2 Transformación de variables numéricas**

Como no es posible eliminar filas o columnas con valores faltantes, ya que se pierde mucha información. Procederemos a modificar los registros numéricos que presentan problemas y a imputar datos para completar los valores faltantes


Ahora vamos a editar los datos que sobrepasan el 100%, y los acotaremos al valor de 100% puesto que no tenemos más opciones para reescribir los datos. Esto con el objetivo de reducir los errores que se puedan obtener al analizar los datos.

Consideramos que este arreglo se realiza unicamente para las variables numéricas con valores porcentuales, las demás no se tendrán en cuenta aquí.

```python
df.info()
```

```python
cols = list(df.columns)
df[cols[3:]] = df[cols[3:]].clip(0, 100)
```

Ya con los valores delimitados, modificamos el dataframe, reemplazando cada valor faltante por el promedio de los datos de la variable de cada municipio por departamento. Por ejemplo, si el municipio Pamplona tiene un valor faltante en la columna 'COBERTURA_NETA', se reemplaza este valor faltante por el promedio de los valores de esta columna de los demás municipios del mismo departamento 'NORTE DE SANTANDER'.

```python
for col in df.columns[3:]:
    df[col] = df.groupby('DEPARTAMENTO')[col].transform(lambda x: x.fillna(x.mean()))

df.info()
```

### **3.1.3 Transformación variable Texto a Entero**

### **4.4.1 Prueba de hipotesis en la cobertura neta de educación primaria entre municipios grandes y pequeños.**



* Se considera municipio grande, aquellos municipios cuya poblacion entre 5 a 16 años es igual o superior a 200000 habitantes.
* Se considera municipio pequeño, aquellos municipios cuya poblacion entre 5 a 16 años es inferior a 200000 habitantes

**Hipotesis a Evaluar:**
* H0 =  No hay diferencia en la cobertura neta de educación primaria entre municipios de diferentes tamaños.
* H1=  La cobertura neta de educación primaria es diferente en municipios de diferentes tamaños.

```python
cobertura_mun_grande = municipios_grandes['COBERTURA_NETA_PRIMARIA']
cobertura_mun_peq = municipios_peq['COBERTURA_NETA_PRIMARIA']
```

Primero verificamos que los datos siguen una distribución aproximadamente normal. Para esto utilizamos la prueba del gráfico Q-Q.

```python
plt.figure(dpi = 110)
stats.probplot(cobertura_mun_grande, plot= plt)
plt.title('Gráfico Q-Q de la cobertura de municipios grandes')
plt.show()
```

```python
plt.figure(dpi = 110) # Mejoramos la visibilidad de la gráfica.
stats.probplot(cobertura_mun_peq, plot= plt)
plt.title('Gráfico Q-Q de la cobertura de municipios pequeños')
plt.show()
```

```python
t_statistic, p_value = stats.ttest_ind(cobertura_mun_grande, cobertura_mun_peq)

print(f'Cobertura promedio municipios grandes: {cobertura_mun_grande.mean()}, Cobertura promedio municipios pequeños: {cobertura_mun_peq.mean()}')
# Imprimir resultados
print("t-statistic:", t_statistic)
print("p-value:", p_value)

# Interpretar los resultados
alpha = 0.05  # nivel de confianza
if p_value > alpha:
    print("No se puede rechazar la hipótesis nula. No hay evidencia suficiente para afirmar que la cobertura neta de educación primaria es diferente en municipios de diferentes tamaños.")
else:
    print("Se rechaza la hipótesis nula. La cobertura neta de educación primaria es diferente en municipios de diferentes tamaños.")
```

Podemos afirmar con un nivel de confianza de $95$%, que hay una diferencia significativa entre la cobertura neta de eduación primaria entre municipios grandes y pequeños. Concretamente, los munipios grandes poseen mayor cobertura promedio que los municipios pequeños
